```python
#Load libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns   
```

```python
df = pd.read_csv(r'netflix_titles.csv')
df
```

```python
#Checking shape of the dataframe
df.shape
```

```python
#Checking for missing values
df.isnull().sum()
```

```python


#filling director missing values with 'Unknown'
df['director'] = df['director'].fillna('Unknown')
#filling cast missing values with 'Unknown'
df['cast'] = df['cast'].fillna('Unknown')
#filling country missing values with 'Unknown'
df['country'] = df['country'].fillna('Unknown')
#filing date_added missing values with 'NaT'
df['date_added'] = df['date_added'].fillna('NaT')
#filling rating missing values with 'Not Rated'
df['rating'] = df['rating'].fillna('NR')
#filling duration missing values with 'Unknown'
df['duration'] = df['duration'].fillna('Unknown')


df
```

```python
#Again Checking for missing values
df.isnull().sum()
```

Data visualization


Movies vs TV Shows Count

```python
sns.countplot(x='type', data=df, palette='Set2')
plt.title('Movies vs TV Shows Count')
plt.show()
```

```python
#Top 10 Countries Producing Netflix Content
top_countries=df['country'].value_counts().head(10)
sns.barplot(x=top_countries.values, y=top_countries.index, palette='cool')
plt.title('Top 10 Countries Producing Netflix Content')
plt.xlabel('Number of Titles')
plt.show()
```

```python
#Content Ratings Distribution
sns.countplot(y='rating',data=df, order=df['rating'].value_counts().index, palette='magma')
plt.title('Distribution of Content Ratings ')
plt.show()
```

```python
#Content Released Over the Years
df['release_year'].value_counts().sort_index().plot(kind='line', figsize=(10,6))
plt.title('Content Release Trend Over the Years')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

```python
#Most Common Genres
from collections import Counter

genres = ','.join(df['listed_in']).split(',')
genre_count = Counter([g.strip() for g in genres])
top_genres = dict(sorted(genre_count.items(), key=lambda x: x[1], reverse=True)[:10])

sns.barplot(x=list(top_genres.values()), y=list(top_genres.keys()), palette='viridis')
plt.title('Top 10 Genres on Netflix')
plt.xlabel('Count')
plt.show()
```
